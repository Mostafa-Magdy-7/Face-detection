```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
img22=cv2.imread("//content/images2.jpg")
cv2_imshow(img22)
```

```python
img22 = cv2.imread("/content/images2.jpg", cv2.IMREAD_GRAYSCALE)

laplacian = cv2.Laplacian(img22, cv2.CV_64F)

laplacian = cv2.convertScaleAbs(laplacian)

plt.subplot(1, 2, 1)
plt.title("Original")
plt.imshow(img22, cmap='gray')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title("Laplacian Filter")
plt.imshow(laplacian, cmap='gray')
plt.axis('off')

plt.show()
```
